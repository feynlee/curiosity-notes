```python
#| eval: false
from pathlib import Path

# %%
def generate_page_for_dir(d):
   """Generate the gallery page for a directory"""
   pics = sorted([pic for pic in d.iterdir() if pic.is_file() and pic.name.lower().endswith('.png')])
   title = d.name.split('-')[-1]
   title = " ".join([w.capitalize() for w in title.split()])
   entries = [
f"""---
title: {title}
format:
   html:
      page-layout: article
filters:
   - lightbox
lightbox: auto
toc: false
---

::::{{.grid}}
"""
   ]

   # make tmp folder to hold renamed pics
   tmp = Path(f"{str(d)}/tmp")
   tmp.mkdir(exist_ok=True, parents=True)

   # rename pics
   for i, pic in enumerate(pics):
      filename = f'{d.name}-{i}.png'
      new_path = f'{str(tmp)}/{filename}'
      pic.rename(new_path)

      # append to listing page
      entries.append(
   f""":::{{.g-col-6 .g-col-md-4}}
![]({tmp.parent.name}/{filename}){{group="my-gallery"}}
:::
"""
         )

   # move pics out of tmp
   for pic in tmp.iterdir():
      new_path = str(pic).replace('tmp/', '')
      pic.rename(new_path)

   # remove tmp folder
   tmp.rmdir()

   # close div
   entries.append("::::")

   # write to file
   art_post = '\n'.join(entries)
   with open(f"{str(d.parent)}/{d.name}.md", "w") as f:
      f.write(art_post)

# %%
p = Path('AI_art/disco_diffusion')

# %%
# dirs = [d for d in p.iterdir() if d.is_dir()]
# for d in dirs:
   # generate_page_for_dir(d)

# %%
d = p/"02-cherry_blossom"
generate_page_for_dir(d)
```
